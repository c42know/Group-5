```python
# Import dependencies
import tweepy
import pandas as pd
import matplotlib.pyplot as plt
import json
import time
import seaborn as sb
import numpy as np
#import datetime
from datetime import datetime
from pprint import pprint
from config import consumer_key, consumer_secret, access_token, access_token_secret
```

```python
# Import and Initialize Sentiment Analyzer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
# Twitter credentials and APi authentification
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

```python
# List to hold results
sentiment = []
# List to hold results
results_list = []

# Target Search Term
target_hashtags = ['#UBER','#LYFT']

# Create array to hold data of tweets
tweet_user = []
tweet_times = []
tweet_texts = []
tweet_location = []

#variables for holding sentiments
compound_list = []
positive_list = []
negative_list = []
neutral_list = []
counter_list = []
target_list = []

#Cities - Washington DC, San Francisco, NYC, CHI, ATL, Las Vegas--IN ORDER#
coordinates = ('38.9072,-77.0369', '37.773972,-122.431297', '40.7128,74.0060', \
              '41.8781,-87.6298','33.7490,-84.3880', '36.1699,-115.1398')
```

```python

# Loop through each hashtag
for hashtag in target_hashtags:
    #set counter
    counter = 1
    print(hashtag)
    
    #loop through coordinates
    for coord in coordinates:

        # Get 100 tweets on targetted tag
        public_tweets = api.search(hashtag, count=100,geocode=coord+",7mi")
        
        print(coord)

        # Loop through all tweets
        for tweet in public_tweets["statuses"]:

            #----TEST---PRINT-LOCATION-----#
            #pprint(tweet['user']['location'])

            # Store all tweet data into the array
            tweet_times.append(tweet["created_at"])
            tweet_texts.append(tweet['text'])
            target_list.append(hashtag)
            tweet_location.append(tweet['user']['location'])
            #----TEST PRINT TWEET----#
            #print(tweet['text'])

            # Run Vader Analysis on each tweet
            results = analyzer.polarity_scores(tweet["text"])
            compound = results["compound"]
            pos = results["pos"]
            neu = results["neu"]
            neg = results["neg"]

            # Add each value to the appropriate list
            compound_list.append(compound)
            positive_list.append(pos)
            negative_list.append(neg)
            neutral_list.append(neu)
            counter_list.append(counter)
            

            # Store the data in dictionary
            sentiment = {
                "User": target_list,
                "Tweet": tweet_texts,
                "Date": tweet_times,
                "Compound": compound_list,
                "Positive": positive_list,
                "Neutral": negative_list,
                "Negative": neutral_list,
                "Tweets Ago": counter_list,
                "Location": tweet_location
            }
            #add to counter
            #print(target)
            counter += 1

#df = pd.DataFrame(sentiment)
sentiments_hash= pd.DataFrame(sentiment)
sentiments_hash.head(100)
```
